### Distribución uniforme 

Dado un espacio muestra finito $S = \{ x_1,x_2,\cdots,x_k\}$ decimos que una distribución es uniforme en $S$ si todos sus valores son igualmente probables. Si la distribución de una variable aleatoria definida en $S$ es uniforme, decimos que $X$ tiene distribución uniforme. Así, si $f$ es la función de masa de probabilidad, entonces
$$ f(x_i) = \frac{1}{k}, $$
para cualquier valores $x_i \in S$. 

**Ejemplo:** Considere los lanzamientos de una dado (que no esté cargado). En este caso, cada uno de los resultados $S=\{1,2,3,4,5,6\}$ son igualmente probables con valor $1/6$.

En este recurso usaremos la librería `scipy` para estudiar las distribuciones de probabilidad.

### Distribución de Bernoulli

Un experimento con dos posibles resultados: 
- un éxito con probabilidad $p$ y denotado por  1,
- un fracaso con probabilidad $q=1-p$ y denotado por 0,

se llama experimento de Bernoulli. Si $X$ representa los resultados de un experimento de Bernoulli se dice que $X$ tiene una distribución de Bernoulli. Decimos que $X$ se distribuye $Bernoulli(p)$.

**Ejemplo:** Considere el lanzamiento de una moneda cargada, de tal forma que obtener una cara tiene el doble de probabilidad de un sello. En este caso, si una cara es un éxito decimos que $X$ se distribuye $Bernoulli(p)$.

- Determine el valor de $p$

```python
# Módulo para utilizar herramientas estadísticas: distribuciones, estadísticas descriptivas, test estadísticos, etc.
# https://docs.scipy.org/doc/scipy/reference/stats.html

from scipy.stats import bernoulli #importamos bernoulli
```

```python
print(ber.pmf(0)) #masa de probabilidad de 0
print(ber.pmf(1)) 
print(ber.cdf(1))
print(ber.sf(0))
```

### Distribución binomial 

Cuando se repiten *n* experimentos independientes de *Bernoulli(p)* tenemos un experimento binomial.

- Consiste en una secuencia de *n* ensayos idénticos.
- Dos resultados son posibles en cada ensayo: éxito o fracaso.
- Las probabilidades no cambian de un ensayo a otro.
- Los ensayos son independientes: el resultado de un ensayo no afecta el otro.

Si $X$ representa los resultados de un experimento binomial decimos que $X$ tiene una distribución *binomial(n,p)*.

En este caso la función de masa de probabilidad $f(k)$ representa la probabilidad de tener $k$ éxitos en $n$ experimentos de *Bernoulli(p)*.

$$f(k)=\left\{\begin{matrix}\\ \binom{n}{k} p^k(1-p)^{n-k}, & k=0, 1, 2, ..., n \\\\ 0 & \text{en otro caso} \\\\ \end{matrix}\right.$$

**Ejemplo:** La moneda cargada del ejemplo anterior se lanza al aire 10 veces. 
- ¿Cuál será la probabilidad de obtener 7 caras?
- ¿Cuál será la probabilidad de obtener 5 caras o menos?
- ¿Cuál la probabilidad de obtener al menos 3 caras?

### Distribución de Poisson

Si $X$ es la variable aleatoria que representa el número de eventos aleatorios independientes que suceden a una rapidez constante, $\lambda$, en el tiempo o espacio, se dice que $X$ tiene una distribución de Poisson($\lambda$).

- Es útil para estudiar el número de ocurrencias de un evento durante un intervalo específico de tiempo o espacio.
- La probabilidad de ocurrencia del evento es la misma para dos intervalos de igual longitud.
- La ocurrencia o no de un evento en cualquier intervalo es independiente de la ocurrencia o no en cualquier otro intervalo.



**Ejemplo:** Se sabe que 10 es el número promedio de camiones que llegan por día a un puerto. Las instalaciones del puerto pueden atender, máximo, 15 camiones en un día. ¿Cuál es la probabilidad de que en un determinado día se tengan que regresar algunos camiones?


```python
from scipy.stats import poisson
```

```python
##Definimos la variable con lambda=10
poi = poisson(10)
```

una vez definida la variables aleatoria *Poisson(10)* podemos contestar a la pregunta anterior y ver la función de masa de probabilidad para la distribución.

```python
print(poi.pmf(8)) #probabilidad lleguen 8 en un día
print(poi.cdf(10)) #Prob lleguen 10 o menos
print(poi.sf(15)) #Respuesta a la pregunta 
```

Tenemos un 0.0487 de probabilidad de que lleguen más de 15 camiones. Veamos la masa de probabilidad.

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(poi.ppf(0.001), poi.ppf(0.99)+4)
ax.plot(x, poi.pmf(x), 'bo', ms=8, label='Función masa de probabilidad')
ax.vlines(x, 0, poi.pmf(x), colors='b', lw=5, alpha=0.5)
ax.set_title('Función de masa de probabilidad Poisson')
ax.legend(loc='best',fontsize=12)
```

Finalmente, cuál será el número de camiones esperado, por día, y su varianza asociada.

```python
## Poi es Poisson(10)
print(poi.mean())
print(poi.var())
print(poi.std())
```

# Distribuciones continuas de probabilidad
Ahora estudiaremos algunas distribuciones de probabilidad continuas que aparecen frecuentemente en problemas aplicados de la estadística.

## Distribución normal 

De entre todas la distribuciones de probabilidad, sin lugar a dudas, la más importante y la que se usa con más frecuencia en la aplicaciones es la **distribución normal**. Diremos que una variable aleatoria $X$ se distribuye normal si tiene función de densidad 

$$ n(x,\mu,\sigma) = \frac{e^{-\frac{1}{2}\frac{(x-\mu )^2}{\sigma^2}}}{\sqrt{2\pi}\sigma} $$
donde $\mu$ y $\sigma$ son parámetros. 

```python
from scipy.stats import norm
```

Grafiquemos la función de densidad de probabilidad para visualizar algunas de sus propiedades. En la siguiente gráfica veremos el efecto del parámetro $\mu$ sobre la densidad de la curva.

```python
#Graficamos distintas densidades
t = np.linspace(-4,4,150)
plt.plot(t,norm.pdf(t),label='$\mu =0$,$\sigma =1$')
# plt.plot(t,norm.pdf(t,loc=2),label='$\mu=2$,$\sigma=1$')
# plt.plot(t,norm.pdf(t,loc=-2),label='$\mu=-2$,$\sigma=1$')
plt.legend(fontsize=16)
plt.title('Normales con diferentes parámetros $\mu$')
plt.show()
```

Veamos ahora el efecto del parámetro $\sigma$ sobre la gráfica de la densidad normal. Para esto, grafiquemos varias curvas con las misma $\mu$ y distintos parámetros $\sigma$.

```python
t = np.linspace(norm.ppf(0.01)-4,norm.ppf(0.99)+4,100)
plt.plot(t,norm.pdf(t),label='$\mu =0$,$\sigma =1$')
plt.plot(t,norm.pdf(t,scale=0.5),label='$\mu=0$,$\sigma=0.5$')
plt.plot(t,norm.pdf(t,scale=2),label='$\mu=0$,$\sigma=2$')
plt.legend(fontsize=16)
plt.title('Normales con diferentes parámetros $\sigma$')
plt.show()
```

Después de ver la influencia de los parámetros en la gráfica de la densidad normal es importante enunciar el siguiente teorema.

**Teorema:** Si $X$ es una distribución normal con parámetros $\mu$ y $\sigma$ entonces 

$$E(X)=\mu\, \text{ y } Var(X)=\sigma^2$$

La demostración del teorema implica algunos artificios de cálculo que omitiremos en este recurso. En lugar de eso, podemos definir una variable aleatoria y preguntar por su media y su varianza.

```python
X = norm(loc=12,scale=3) #normal mu=12, sigma=3
```

```python
print(X.mean()) #¿Cuál será el valor esperado?
print(X.var()) #¿cuál su varianza?
print(X.std())
```

Ademas del teorema anterior es importante hacer notar varias características que se pueden deducir directamente de la gráfica de la densidad:
1. La distribución normal (o gaussiana) es simétrica respecto a la media.
2. Se puede mostrar que los puntos de inflexión de la curva ocurren en $\mu+\sigma$ y $\mu-\sigma$.
3. Cuando $x\to \pm \infty$ la curva tiene a 0.
4. El área bajo la curva normal es igual a 1.

Una manera típica de visualizar la distribución gaussiana es usanda un histograma de frecuencias relativas. En la siguiente celda simularemos varios valores de la distribución normal y graficaremos su histrograma.

```python
sim = X.rvs(1000) #Simulamos 1000 muestras de la normal
plt.hist(sim,density=True,bins=32)
plt.show()
```

Apropósito, recuerden que las áreas bajo las densidades se interpretan como las probabilidades. 

**Ejemplo:** Si $X$ es una distribución normal con $\mu=0$ y $\sigma=1$ halle 
1. $P(X>1.84)$
2. $P(-1.97 < X < 0.86)$

La gráfica siguiente muestra la probabilidad (es decir, el área) que nos interesa.

```python
t = np.linspace(-4,4,100)
plt.plot(t,norm.pdf(t),label='$\mu =0$,$\sigma =1$')
x=np.linspace(-1.97,0.86,150)
plt.fill_between(x,norm.pdf(x),color='b',alpha=0.3)
plt.legend()
plt.title('$P(-1.97<X<0.86)$ :Área bajo la curva normal')
plt.show()
```

```python
x = norm(loc=0,scale=1)
#print(x.sf(1.84)) #Respuesta a 1
#print(1-x.cdf(1.84)) #Respuesta a 1
## Respuesta a 2
print(x.cdf(0.86)-x.cdf(-1.97))
```

Sabemos que en general, una probabilidad como la planteada en el ejercicio anterior se puede obtener usando la función de distribución acumulada. Para el caso de una distribución normal estándar se tiene que 

$$ P(X\leq x)=F(x) = \displaystyle\int_{-\infty}^x \frac{e^{-t^2/2}}{\sqrt{2\pi}}dt $$

Sabemos que usando la función de distribución acumulada podemos obtener la respuesta al problema anterior y también preguntar los los cuantiles de la distribución ¿Cuál será el cuantil $q_{0.25}$ (es decir, el límite inferior del segundo cuartil)?

```python
t = np.linspace(norm.ppf(0.01)-4,norm.ppf(0.99)+4,100)
plt.plot(t,norm.cdf(t),label='$\mu =0$,$\sigma =1$')
x = np.linspace(-4,norm.ppf(0.25),50)
plt.fill_between(x,norm.cdf(x),color='b',alpha=0.3)
plt.legend()
plt.title('Función de distribución acumulada $N(0,1)$')
plt.show()
```

Ahora señalamos una propiedad muy importante que se usa ampliamente en las aplicaciones de la distribución normal

**Definición:** Si una variable aleatoria $X$ se distribuye normal con medio 0 y varianza 1, se dice que $X$ es una normal estándar y lo denotaremos $X\sim N(0,1)$.

**Nota:** Si $X\sim N(\mu,\sigma^2) $ entonces la variable 
$$Z=\frac{X-\mu}{\sigma} \sim N(0,1),$$
es decir, $Z$ es una normal estándar. Esta propiedad y ciertas tabulaciones para la función de distribución acumulada se usan ampliamente en los textos de estadística, ya que la integración de la densidad gaussiana no se puede hallar por los métodos analíticos.

Otra propiedad que se usará ampliamente en inferencia es el hecho de que el 95% del área bajo la densidad gaussiana se encuentra en un intervalo que mide 4 desviaciones estándar y está centrado en la media. Este fenómeno se ilustra en la siguientes celda

```python
t = np.linspace(-4,4,200)
plt.plot(t,norm.pdf(t),label='$\mu =0$,$\sigma =1$')
x=np.linspace(-2,2,100)
plt.fill_between(x,norm.pdf(x),color='b',alpha=0.3)
plt.legend()
plt.title('95% del área en el intervalo [-2,2]')
plt.show()
```

## Distribución log-normal

Sea $X$ una v.a. no negativa y $Y=\ln(X)$. Si $Y\sim N(\mu, \sigma^2)$, entonces decimos que $X$ tiene distribución log-normal y escribimos $$X\sim LogN(\mu, \sigma^2)$$

La función densidad de probabilidad de una v.a. log-normal esta dada por:

$$f(x)=\left\{\begin{matrix} \dfrac{1}{\sigma \sqrt{2\pi} x}e^{\frac{-(\ln(x) - \mu)^2}{2\sigma^2}}, & x > 0 \\\\ 0 & \text{en otro caso} \\ \end{matrix}\right.$$

```python
from scipy.stats import lognorm
```

```python
#Graficamos distintas densidades
dist_lognorm1=lognorm(1,loc=0)
dist_lognorm2=lognorm(0.5,loc=0)
dist_lognorm3=lognorm(1.5,loc=0)
t = np.linspace(0,10,200)
plt.plot(t,dist_lognorm2.pdf(t),label='$\mu =0$, $\sigma =0.5$')
plt.plot(t,dist_lognorm1.pdf(t),label='$\mu =0$, $\sigma =1$')
plt.plot(t,dist_lognorm3.pdf(t),label='$\mu =0$, $\sigma =1.5$')
plt.legend(fontsize=16)
plt.title('Log-normales con diferentes parámetros $\sigma$')
plt.show()
```

```python
#Graficamos distintas densidades
dist_lognorm1=lognorm(1,loc=0)
dist_lognorm2=lognorm(1,loc=1)
dist_lognorm3=lognorm(1,loc=2)
t = np.linspace(0,10,200)
plt.plot(t,dist_lognorm1.pdf(t),label='$\mu =0$, $\sigma =1$')
plt.plot(t,dist_lognorm2.pdf(t),label='$\mu =1$, $\sigma =1$')
plt.plot(t,dist_lognorm3.pdf(t),label='$\mu =2$, $\sigma =1$')
plt.legend(fontsize=16)
plt.title('Log-normales con diferentes parámetros $\sigma$')
plt.show()
```

**Teorema**: Si $X$ se distribuye log-normal con parámetros $\mu$ y $\sigma$, entonces

$$\mathbb{E}(X) = e^{\mu + \frac{\sigma^2}{2}}, \ \mathbb{V}\text{ar}(X) = e^{\sigma^2 - 1}(e^{2\mu + \sigma^2})$$

```python
# Ejemplo

X = lognorm(1, loc=0)
```

```python
print(X.mean())
print(X.var())
print(X.std())
```
